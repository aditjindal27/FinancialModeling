```python
from polygon import RESTClient
```

```python
KEY = ""
client = RESTClient(api_key=KEY)
```

```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
stocks_list = pd.read_csv("s&p 500 stocks.csv")
```

```python
stocks = stocks_list["Symbol"].T.tolist()
```

```python
stocks
```

```python
#Get last 1 year data for stock
def getDailyClose(ticker):
    obj =  client.get_aggs(ticker=ticker, multiplier = 1, timespan="day", from_ = "2022-01-01", to="2023-01-01")
    vals = {}
    for i in range(len(obj)-1, -1, -1):
      vals[obj[i].timestamp] = obj[i].close
      
    return pd.DataFrame.from_dict(vals, orient='index', columns=[ticker])
```

```python
func = lambda x: datetime.fromtimestamp(x/1000.0).strftime("%Y-%m-%d")
```

```python
table = None
for i in range(len(stocks)):
  if i == 0:
    table = getDailyClose(stocks[i])
  else:
    if stocks[i] not in ["GEHC"]:
      table = table.join(getDailyClose(stocks[i]), how='outer')
table.index = table.index.map(func)
```

```python
table = table.fillna(0)
```

```python
table
```

```python
table.index.names = ["Date"]
```

```python
table = table.reset_index()
```

```python
table.to_csv("prices_S&P.csv")
```
